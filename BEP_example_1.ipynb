```python
%matplotlib inline
import sympy as sm
from sympy.physics.continuum_mechanics.beam import Beam
```

```python
E = sm.Symbol('E')
I = sm.Symbol('I')
b = Beam(10, E, I)
b.apply_support(0, type="pin")
b.apply_support(10, type="pin")
b.apply_load(-20, 5, -1)
R_0, R_10 = sm.symbols('R_0, R_10')
b.solve_for_reaction_loads(R_0, R_10)
b.reaction_loads
```

```python
b.plot_bending_moment();
```
